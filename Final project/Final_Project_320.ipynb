# Introduction and Motivation:

Our goal is to examine how citizenship, gender, occupation, state, and family composition influence working hours, in order to enhance our understanding of participants' labor market behavior. In particular, we are interested in examining how these variables affect the number of hours worked and if specific populations are more inclined to longer or shorter working hours. This issue is set within the global context of contemporary labor markets characterized by the interplay of economic, sociocultural, and political factors that dictate the work – life balance. Dismissing the importance of work hours is economically counterproductive, socially conservative, and regressive. In addition, work hours differ among certain demographic groups due to the availability of employment, family responsibilities, social attitudes toward gender, and immigration rules. Through this analysis, we strive to offer insights for formulating policy geared towards creating discrimination-free laws in the labor market, fostering inclusive workplaces, and making family and career choices. Advances in understanding work participation disparities, workforce policies, and economic equity will require grappling with these dynamics. To substantiate this inquiry are studies of the labor market, employment indicators, and economic theories of social inequality and work-life balance, which help contextualize and validate the findings.

# Literature Review

## Children and Careers: How Family Size Affects Parents’ Labor Market Outcomes in the Long Run
- **Author:** Sara Cools, Simen Markussen, Marte Strom
- **[Link to article](https://pmc.ncbi.nlm.nih.gov/articles/PMC5624987/)**  
The correlation between family size and the workforce, mainly weekly hours worked, has received considerable focus. Cools et al. (2017) use instrumental variables to analyze this and demonstrate that family size strongly and negatively affects mothers’ weekly working hours while fathers’ hours remain primarily unchanged. In essence, they argue that each additional child decreases mothers’ labor supply by, on average, 3-5 hours a week, with more pronounced effects during the early years following childbirth and enduring impacts in the scenario of highly educated women. As described, the study illustrates Becker’s well-known economic concepts, a posteriori validation of the econometric assumptions of the Beckerian household model, and a quantity-quality trade-off, which states that an increase in market work necessitated in response to increased childcare duties. Additionally, robust gaps in gender differences in the labor supply perpetuate the maternal employment paradigm, which positions women as principal caregivers. These findings suggest an urgency for more attention to subsidized child care, flexible work hours, active father involvement, and a reduction in the detrimental effects on women’s work hours. The study illustrates the implications of family size on parental participation in the labor market alongside the need for policies on gender inequality in the labor market.

## Gender Inequality, Work Hours, and the Future of Work
- **Authors:** Ariane Hegewisch, M.Phil., and Valerie Lacarte, Ph.D.
- **[Link to report](https://iwpr.org/wp-content/uploads/2020/07/C486_FOW-Work-Hours-Report.pdf)**  
The Institute for Women’s Policy Research in 2020 stipulated that there is an unequal allocation of work opportunities to men and women, which in one way or another influences the number of hours each gender is required to work. Women, in most cases, do not work as many hours as men in a week because of an amalgamation of caregiving duties, social norms, and a lack of accepted domestic help, which women have to shoulder alongside their primary work. Women are more likely than men to work shorter hours, not because they want to, but because they are trying to juggle caregiving responsibilities, which decreases their earnings. But it seems men actively take on more full-time and overtime positions that offer set hours. Thus, their earnings increase substantially compared to women's because established practices confirm that subordinate social classes have to work hard to earn good money, and this is especially true in low-paying jobs within the service sector where women are hopelessly trapped.

## Immigration and work schedules: Theory and evidence
- **Authors:** Timothy N. Bond, Osea Giuntella, Jakub Lonsky
- **[Link to report](https://www.sciencedirect.com/science/article/pii/S0014292122002380)**  
Regardless of how their weekly employment compares to that of native workers, immigrants are known to work more hours per week, especially during the night. In a study conducted by Bond et al. (2023), it was established that immigrants ‘disproportionately’ fill positions featuring longer or irregular hours because of economic necessity, lower disutility costs, and lack of alternative employment options. Consequently, native workers in regions with higher inflows of immigrant workers tend to receive disproportionate daytime shifts. The reasoning is that immigration to an area causes a decline in compensating differentials for night work. Immigrants tend to accept lower wages and work night shifts, which are undesirable. The study shows that a 10 percentage point increase in the share of immigrants in the workforce decreases the probability of natives working night shifts by 0.8 percentage points and increases wages for daytime shifts. The findings highlight that citizenship status is crucial in allocating work schedules, especially non-citizens, who are overrepresented and fill lower-amenity, extended, or irregular-hours jobs. Even as natives enjoy flexible access to standard work schedules, immigrants are compelled to work longer hours in low-paying, labor-intensive, and monotonous job sectors.

- **Authors:** Alice Gibbs
- **[Link to article](https://www.newsweek.com/map-shows-states-people-overworked-1966944)**  
The Newsweek (2024) article analyzes overwork within U.S. states regarding geography and highlights apparent differences in workload, balance of personal life, and leisure time across regions. Louisiana is labeled as the most overworked state, with an average workweek of 36.5 hours, remote work available at only 6.2%, and a high workplace death rate of 6.4 per 100,000 workers. North Dakota, widely considered the hardest working state, reports a near 98% employment rate alongside a 39.7-hour workweek, with one-third of employees not taking vacation days. Alaska has the highest average workweek at 41.6 hours per week. On average, Americans work more than 1,811 hours a year, more than most developed countries. These observations enhance research on the geography of work about state policies determining overwork, work-life balance, and employee health and welfare.

## Occupation and Working Hours (ATUS 2023)
- **Authors:** Bureau of Labor Statistics
- **[Link to article](https://www.bls.gov/news.release/pdf/atus.pdf)**  
The 2023 American Time Use Survey by the U.S. Bureau of Labor Statistics shows variation in working hours by occupation. Construction and extraction workers had the most extended average workday (8.69 hours), followed by production (8.43). Service work was reported to be the shortest at 7.22. Remote work patterns also differed: approximately 49% of management and finance employees worked from home, averaging 5.7 hours remotely, while only 16% of workers in service roles did. Differences by gender also persist; men work 36 minutes more than women daily. These American time-use survey trends show how a person’s occupation and workplace standing elicit differing shifts in the U.S. labor market.

The reason why we chose the following variables for the simple regression model is due to the fact that we believe that the variables listed above are crucial and are heavily correlated to the dependent variable (the working hours per week) in reality.

1. Citizenship
Citizenship status is a key socioeconomic indicator that can affect the types of jobs individuals hold and the conditions under which they work. Non-citizens may face fewer job protections and more economic pressure, leading them to accept longer or less flexible work hours. Including this variable allows us to assess how legal and structural factors tied to immigration status shape labor patterns.

2. Gender
Gender is one of the most widely studied dimensions in labor economics, and prior research consistently shows significant differences in work hours between men and women. These disparities often stem from caregiving responsibilities, occupational segregation, and societal expectations. Including gender helps us quantify how these patterns show up in weekly hours worked.

3. Family Size
Family size is a proxy for household responsibilities and financial obligations. Larger families may push individuals to work more hours to meet household needs, although the effect could vary depending on access to childcare, social support, or cultural norms. We include this variable to test whether people with more dependents actually work longer hours.

4. Region
Geographic region affects employment opportunities, job types, and cost of living. For example, labor markets in the South differ from those in the Northeast in terms of job structure, wages, and hours expectations. By including region, we can account for broad structural differences across U.S. labor markets that may influence how much people work.

5. Occupation
Occupation captures the type of work someone does, which directly influences work hours. Some professions (like healthcare, construction, or management) typically demand longer schedules, overtime, or irregular hours. Including occupation helps us control for job-specific effects on work intensity and better isolate the impact of other variables like gender or citizenship.

6. Education
Educational attainment is a strong predictor of labor market outcomes, including job type, stability, and compensation. Individuals with higher levels of education may have access to more flexible or higher-paying jobs, which can affect their weekly working hours. Conversely, those with lower educational attainment may work longer or irregular hours to compensate for lower wages. Including this variable allows us to control for human capital differences and better isolate the effects of other socioeconomic factors like occupation or region.

## Results

# 1. Production, Transportation, and Material Moving

Region: Workers in the South worked about 0.38 more hours per week than those in the Midwest (p = 0.044), while those in the Northeast and West worked significantly fewer hours (−1.49 and −0.33, respectively), with the Northeast effect being statistically significant (p < 0.001).

Education: No educational category was statistically significant compared to the college-educated reference group.

Gender: Females worked approximately 3.89 fewer hours per week than males (p < 0.001).

Citizenship and Family Size: Both had small and statistically insignificant effects on hours worked.

# 2. Service
Region: Individuals in the South and West worked 1.63 and 1.17 more hours, respectively, than those in the Midwest, with both effects being statistically significant (p < 0.001).

Education: Workers with only a high school diploma worked 1.59 fewer hours (p < 0.001), and those without a high school diploma worked 2.89 fewer hours per week (p < 0.001), compared to those with a college degree.

Gender: Females worked 4.08 fewer hours per week than males (p < 0.001).

Citizenship: Citizens worked 0.24 fewer hours per week on average (p = 0.015).

Family Size: Had a small and marginally significant positive effect (p = 0.061).

# 3. Natural Resources, Construction, and Maintenance
Region: Workers in the South worked 1.31 more hours per week than those in the Midwest (p < 0.001); other regional differences were not statistically significant.

Education: High school and sub-high school educational levels were associated with 0.66 and 1.34 fewer hours respectively (both p < 0.001) compared to college-educated workers.

Gender: Female workers in this sector worked about 4.32 fewer hours (p < 0.001).

Citizenship and Family Size: Both had positive and significant effects (citizenship: +0.40 hours, p < 0.001; family size: +0.07 hours, p = 0.015).

# 4. Sales and Office
Region: The only significant regional difference was found in the South, where workers reported 0.25 more hours than their Midwest counterparts (p = 0.040).

Education: Both lower education categories were associated with fewer hours worked (−0.76 and −0.38), although only the former was statistically significant (p < 0.001).

Gender: Female workers worked approximately 2.04 fewer hours per week (p < 0.001).

Citizenship: Citizens worked 0.44 fewer hours (p < 0.001).

Family Size: Larger family sizes were associated with more hours worked (+0.17, p < 0.001).

# 5. Management, Business, Science, and Arts
Region: Workers in the Northeast and West worked significantly fewer hours (−1.89 and −1.55, both p < 0.001) compared to the Midwest.

Education: Only the “No High School Diploma” group showed a significant effect (−0.83 hours, p = 0.001).

Gender: Females worked 2.68 fewer hours (p < 0.001).

Citizenship: Positive but marginally significant (p = 0.090).

Family Size: Larger family size was associated with fewer hours worked (−0.11, p < 0.001).

# Conclusion 
	Our study explored how weekly hours worked vary across demographic characteristics and regions within the U.S. labor force, explicitly focusing on differences across major occupational categories. Using OLS regression models applied separately to five broad occupation groups, we examined the influence of gender, citizenship status, education level, region, and family size on labor supply, measured in hours worked per week.
	Across all occupations, gender emerged as the most consistent and significant factor: individuals identified as female worked approximately 2 to 4 fewer hours per week than their male counterparts, even after adjusting for education, region, family size, and citizenship. This persistent gap highlights structural inequalities and the ongoing influence of gender roles in shaping work hours. 
	Citizenship effects varied by occupation. While it was positively associated with hours worked in some fields, it was negatively associated or insignificant in others, suggesting that job access and work-hour expectations differ meaningfully across sectors.	
	Education levels also showed nonuniform effects. In some categories, like Sales and Office or Service, having less than a high school diploma was associated with longer hours. In contrast, in others, the relationship was less clear or insignificant.	
	Regional differences were most pronounced when comparing the South and West to the Midwest (our omitted baseline), with individuals in the South generally working more hours across several occupations. However, the effect sizes varied, indicating regional labor market structures interact differently with job types.
	Although the R² values were relatively low across models, with adjusted R² ranging from 0.011 to 0.033, these results still provide meaningful insight into labor supply patterns. Our models are not designed to capture all drivers of hours worked but to identify consistent patterns across observable characteristics.
